```python
Product_name = []
prices = []

for i in range(2,11):
    url = 'https://www.flipkart.com/search?q=watch&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page='+str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text,'lxml' )
    # print(soup)
    np = soup.find('a',class_ = '_9QVEpD').get('href')
    comp_next_page = 'https://www.flipkart.com' + np

    name = soup.find_all('a',class_='WKTcLC')
    
    for i in name:
        Pro_name = i.text
        Product_name.append(Pro_name)

    price = soup.find_all('div',class_='Nx9bqj')
    filtered_elements = [div for div in price if div['class'] == ['Nx9bqj']]

    
    for i in filtered_elements:
        # print(i.text)
        pro_price = i.text
        prices.append(pro_price)
```

```python
Product_name
```

```python
print(len(Product_name))
print(len(prices))
```

```python
df = pd.DataFrame({"Product_name":Product_name,'Prices':prices})
df.to_excel('flipcart360.xlsx',index=False)
```
