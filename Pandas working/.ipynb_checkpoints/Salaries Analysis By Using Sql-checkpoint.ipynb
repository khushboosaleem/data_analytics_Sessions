```python
# import important libraries
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect("database.sqlite")
conn
```

```python
def sql_query(q):
    return pd.read_sql_query(q, conn)
```

```python
sql_query("select * from sqlite_master")
```

```python
sql_query("select * from sqlite_master where type = 'table'")
```

### ***1. Show All The Columns And Rows in the table***

```python
q = "select * from Salaries"
sql_query(q)
```

### ***2. Show Only the EmployeeName and JobTitle columns.***

```python
q = 'select EmployeeName, JobTitle from salaries'
sql_query(q)
```

### ***3. Show The Number Of Employees In The Table***

```python
q = 'select count(*) from salaries '
sql_query(q)
```

### ***4. Show The Unique Job Titles In The Table***

```python
q = 'select distinct JobTitle from salaries'
sql_query(q)
```

### ***5. Show The JobTitle and Overtime Pay for All Employees With Overtime Pay greater than 50000***

```python
q = '''select JobTitle, OvertimePay from salaries
      where OvertimePay	> 50000 AND OvertimePay is not null
      AND OvertimePay != 'Not Provided'
'''
sql_query(q)
```

## ***6. Show the Average BasePay for all Employees.***

```python
q = '''select avg(BasePay) as "Average BasePay" from salaries'''
sql_query(q)
```

## ***7. Show The Top 10 highest Paid Employees***

```python
q = '''select EmployeeName, TotalPay from salaries
    order by TotalPay desc limit 10
'''
sql_query(q)
```

### ***8. Show The Average of BasePay, OverTime, OtherPay for Each Employee***

```python
q = '''SELECT EmployeeName, (BasePay + OvertimePay + OtherPay) / 3 AS AveragePay
        FROM salaries
        WHERE EmployeeName IS NOT NULL AND EmployeeName != 'Not provided'     
'''
sql_query(q)
```

### ***13. Show All Employees With A TotalPay Benefits Value Between 125,000 and 150,000 And A Job Title Containing Word 'Director'***

```python
q = '''select TotalPayBenefits, JobTitle from salaries
    where TotalPayBenefits between 125000 and 150000
    AND 
    JobTitle LIKE "%Director%"
'''
sql_query(q)
```

### ***14. Show All Employees Ordered By Their TotalPay Benefits In Descending Order***

### ***15. Show All Job Titles With An Average BasePay Of Atleast 100,000 And Order Them By The Avg BasePay In Descending Order*** 

```python
q = '''select JobTitle, AVG(BasePay) as 'Avgbasepay' from salaries
    group by JobTitle
    having avg(BasePay) >= 100000
    order by Avgbasepay desc
'''
sql_query(q)
```

### ***16. Updating The BasePay Of All Employees With The Job Title Containing "Manager" By Increasing It By 10%***

```python
def sql_query(q):
    cursor = conn.cursor()  # Create a cursor
    cursor.execute(q)  # Execute the SQL query
    conn.commit()  # Commit the transaction
    cursor.close()  # Close the cursor
```

```python
q = '''update salaries
    set BasePay = BasePay * 1.1
    where JobTitle like '%Manager%'
'''
sql_query(q)
```
