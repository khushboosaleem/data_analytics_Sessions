## ***17. Replace income value [<=50k with 0 and >50k with 1]***

```python
data["income"].unique()
```

```python
sns.countplot(x='income',data=data,palette='viridis',hue='gender')
plt.show()
```

```python
def encoded_income(sal):
    if sal == '<=50K':
        return 0
    else:
        return 1 
```

```python
encoded_income('<=50k')
```

```python
data['encoded_salary'] = data['income'].apply(encoded_income)
data.head()
```

```python
data['encoded_salary'].unique()
```

```python
# Another way to replace
data.replace(to_replace = ['<=50K', '>50K'], value = [0,1],inplace=True)
```

```python
data.head()
```
