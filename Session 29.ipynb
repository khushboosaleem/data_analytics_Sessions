```python
# invoice_id, track_id, track_name, media_type name, quantity, unit_price  where invoice_id = 1

q = """select il.invoice_id, t.track_id, t.name track_name, m.name media_name, il.quantity, il.unit_price
        from invoice_line il
        inner join track t
        on 
        il.track_id = t.track_id
        inner join media_type m
        on 
        t.media_type_id = m.media_type_id
        where il.invoice_id == 1

"""

sql_query(q)
```

```python
# invoice_id, track_id, track_name, artist_name, media_type, quantity, unit_price  = where invoice_id = 1

q = """select il.invoice_id, t.track_id, t.name track_name,ar.name artist_name, m.name media_name, il.quantity, il.unit_price
        from invoice_line il
        inner join track t
        on 
        il.track_id = t.track_id
        inner join media_type m
        on 
        t.media_type_id = m.media_type_id
        inner join album al
        on 
        t.album_id = al.album_id
        inner join artist ar
        on
        al.artist_id = ar.artist_id
        
        where invoice_id == 1

"""

sql_query(q)
```

```python
# tract_id, album title, artist name

q = """select t.track_id , al.title album_title, ar.name artist_name
        from track t
        inner join album al
        on
        t.album_id = al.album_id
        inner join artist ar
        on 
        al.artist_id = ar.artist_id

"""
sql_query(q)
```

```python
# no_of_tracks_sale, album_name, artist_name 

q = """select count(il.track_id) No_of_tracks, al.title album_name, ar.name artist_name
        from invoice_line il
        inner join track t
        on
        il.track_id = t.track_id
        inner join album al
        on
        t.album_id = al.album_id
        inner join artist ar
        on
        al.artist_id = ar.artist_id
        group by 2

"""
sql_query(q)
```

```python
# Top 5 track sale 

q = """select count(il.track_id) No_of_tracks_sale, al.title album_name, ar.name artist_name
        from invoice_line il
        inner join track t
        on
        il.track_id = t.track_id
        inner join album al
        on
        t.album_id = al.album_id
        inner join artist ar
        on
        al.artist_id = ar.artist_id
        group by 2
        order by 1 DESC limit 5

"""
sql_query(q)
```

```python
# Top five Customer_name, No_of_invoices, Total_spent

q = """select c.first_name ||" "|| c.last_name Customer_name,
        count(inv.invoice_id) No_of_Invoices,
        sum(inv.total) Total_spen
        from customer c
        inner join invoice inv
        on
        c.customer_id = inv.customer_id
        group by c.customer_id
        order by 3 DESC limit 5

"""
sql_query(q)
```

```python
# Top five Customer_name, No_of_invoices, Total_spent

q = """select c.first_name ||" "|| c.last_name Customer_name,
        count(inv.invoice_id) No_of_Invoices,
        sum(inv.total) Total_spen
        from customer c
        inner join invoice inv
        on
        c.customer_id = inv.customer_id
        group by c.customer_id
        

"""
sql_query(q)
```

```python
# Add new col customer_category 

q = '''SELECT first_name || ' ' || last_name Customer_Name, 
        count(inv.invoice_id) No_Of_Invoices,
        sum(total) Total_Spent,
        Case
            when sum(inv.total) < 40 then "Small Spender"
            when sum(inv.total) > 100 then "Big Spender"
            Else "Regular"
        End as Customer_category
        from customer c
        inner join invoice inv
        on 
        c.customer_id = inv.customer_id
        group by c.customer_id
        
        
'''
sql_query(q)
```

```python
# value_count 
q = '''SELECT first_name || ' ' || last_name Customer_Name, 
        count(inv.invoice_id) No_Of_Invoices,
        sum(total) Total_Spent,
        Case
            when sum(inv.total) < 40 then "Small Spender"
            when sum(inv.total) > 100 then "Big Spender"
            Else "Regular"
        End as Customer_category
        from customer c
        inner join invoice inv
        on 
        c.customer_id = inv.customer_id
        group by c.customer_id
        order by 1
        
'''
sql_query(q)["Customer_category"].value_counts()
```

```python
# customer all col , total invoice
q = """select c.*, sum(inv.total) Total_Spent
        from customer c
        left join invoice inv
        on
        c.customer_id = inv.customer_id
        group by c.customer_id
        having Total_Spent > 90

"""
sql_query(q)
```

```python
# customer all col , Top 3 total invoice
q = """select c.*, sum(inv.total) Total_Spent
        from customer c
        left join invoice inv
        on
        c.customer_id = inv.customer_id
        group by c.customer_id
        having Total_Spent > 90
        order by Total_Spent DESC limit 3
        

"""
sql_query(q)
```

```python
# Create table

q = """create view CUSTOMER_ABV3_90 as 
        select c.*, sum(inv.total) Total_Spent
        from customer c
        left join invoice inv
        on
        c.customer_id = inv.customer_id
        group by c.customer_id
        having Total_Spent > 90
        order by Total_Spent DESC limit 3
        

"""
conn.execute(q)
```
