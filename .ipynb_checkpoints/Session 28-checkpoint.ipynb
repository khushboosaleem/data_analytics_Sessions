```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect('factbook.db')
conn
```

```python
def sql_query(q):
    return pd.read_sql_query(q, conn)
```

```python
q = '''select * from sqlite_master

'''
sql_query(q)
```

```python
q = '''pragma table_info('sqlite_sequence')

'''
sql_query(q)
```

```python
q = '''pragma table_info('facts')

'''
sql_query(q)
```

```python
q = '''pragma table_info('cities')

'''
sql_query(q)
```

### **Show all data from facts table**

```python
q = '''select * from facts

'''
sql_query(q)
```

### **Show all data from cities table**

```python
q = '''select * from cities

'''
sql_query(q)
```

### **Show all data from facts table where name == Pakistan**

```python
q = '''select * from facts
       where name == 'Pakistan'     
'''
sql_query(q)
```

### **Show all data from cities where fact_id == 132**

```python
q = '''select * from cities
    where facts_id == 132
'''
sql_query(q)
```

### **Retrieve all rows where there is a matching record in both the facts and cities tables, using the id column in the facts table and the facts_id column in the cities table as the join condition.**

```python
q = '''select * from facts f  
    inner join cities c
    on
    f.id = c.facts_id
    
'''
sql_query(q)
```

### **Show the country name and city name for entries where the country is 'Pakistan'**

```python
q = '''select f.name Country , c.name City from facts f
        inner join 
        cities c
        on 
        f.id = c.facts_id
        where f.name == 'Pakistan'
'''
sql_query(q)
```

### **find countries that do not have any cities listed in the cities table. The query should return the country name and city name.**

```python
q = '''select f.name Country , c.name from facts f
        left join cities c
        on
        f.id = c.facts_id
        where c.name is null

'''
sql_query(q)
```
