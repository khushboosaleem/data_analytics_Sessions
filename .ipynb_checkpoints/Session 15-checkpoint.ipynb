```python
class Car():
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 1000

    def read_odometer(self):
        print(f"This car has {self.odometer_reading} miles on it.")

    def update_odometer(self, mileage):
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")
        return self.odometer_reading

    def increment_odometer(self, miles):
        self.odometer_reading += miles

    def get_discriptive_name(self):
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name.title()


car1 = Car('audi','A3',2022)

print(car1.get_discriptive_name())
print(car1.read_odometer())
print(car1.update_odometer(2500))
print(car1.read_odometer())
print(car1.increment_odometer(600))
print(car1.read_odometer())
```

```python
class ElectricCar(Car):
    def __init__(self, make, model, year,color):
        super().__init__(make, model, year)
        self.color = color
        self.battery_size = 75

    def describe_battery(self):
        print(f"This car has a {self.battery_size} kWh battery.")

    def car_color(self):
        print(f"This car is {self.color}.")

my_tesla = ElectricCar('tesla', 'model s', 2019,'red')
print(my_tesla.get_discriptive_name())
my_tesla.describe_battery()
my_tesla.car_color()
```

```python
class Employee():
    def __init__(self,name,age,salary):
        self.name = name
        self.age = age
        self.salary = salary

    def get_salary(self):
        print(f"The salary of {self.name} is {self.salary}")

emp1 = Employee("Zeeshan","34",70000)
emp1.get_salary()

class Driver(Employee):
    def __init__(self,name,age,salary):
        super().__init__(name,age,salary)

    def get_salary(self):
        self.salary = self.salary/2
        print(f"The salary of {self.name} is {self.salary}")

emp2 = Driver("Anees","35",70000)
emp2.get_salary()
```

```python
class Shapes():
    def details(self):
        print("This is a shape")
class Circle(Shapes):
    def details(self):
        print("This is a circle")
class Square(Shapes):
    def details(self):
        print("This is a square")
class Rectangle(Shapes):
    def details(self):
        print("This is a rectangle")

shape1 = Shapes()
shape1.details()
shape2 = Circle()
shape2.details()
shape3 = Square()
shape3.details()
shape4 = Rectangle()
shape4.details()
```

```python
class Employee():
    def __init__(self,name,over_time_rate,basic_salary):
        self.name = name
        self.over_time_rate = over_time_rate
        self.basic_salary = basic_salary

    def calculate_salary(self):
        total_overtime_amount = 30 * self.over_time_rate
        total_salary = self.basic_salary + total_overtime_amount
        print(f"The salary of {self.name} is {total_salary}")

    def update_salary(self,new_salary):
      print(f"previous was {self.basic_salary}. updating new salary is {new_salary} ")
      self.basic_salary = new_salary

emp1 = Employee("Zeeshan",400,80000)
emp1.calculate_salary()
emp1.update_salary(100000)
emp1.calculate_salary()
```

```python
class Car():
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 1000

    def get_discriptive_name(self):
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name.title()

    def fill_tank(self):
        print("Filling tank")

    def read_odometer(self):
        print(f"This car has {self.odometer_reading} miles on it.")

    def update_odometer(self, mileage):
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")
        return self.odometer_reading

    def increment_odometer(self, miles):
        self.odometer_reading += miles

car1 = Car('audi','A3',2022)
print(car1.get_discriptive_name())
car1.read_odometer()
print(car1.update_odometer(2500))
car1.read_odometer()
car1.increment_odometer(600)
car1.read_odometer()
car1.fill_tank()
```
